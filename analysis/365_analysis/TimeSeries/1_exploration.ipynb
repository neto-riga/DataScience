## QQ-plot
---
QQ-plots are a fast way to check if the values of our dataset follow a given distribution<br>
To check the density of the data and the values that are more likely to occur.

```python
import scipy.stats
import pylab
```

```python
scipy.stats.probplot(df_cop['spx'], plot = pylab)
pylab.show()
```

The theoretical quantiles can be seen as **how many standard deviations** away from the **mean** these values are.

The redline shows how should the points look if they were normally distributed. In our data we can see that our data is not, due to the left side of our data.

```python
scipy.stats.probplot(df_cop['dax'], plot = pylab)
pylab.show()
```
