Con la misma idea del data frame de los aguacates, haremos un nuevo data frame con los estados como columnas y el salario mínimo como el valor de las filas.

```python
gb = df.groupby("State")
gb.get_group("Alabama").set_index("Year").head(4)
```

Ahora para cada estado

```python
min_wage = pd.DataFrame()

for name, group in df.groupby("State"):
    if min_wage.empty:
        min_wage = group.set_index("Year")[["Department.Of.Labor.Cleaned.Low.Value.2020.Dollars"]].rename(columns={'Department.Of.Labor.Cleaned.Low.Value.2020.Dollars': name})
    else:
        min_wage = min_wage.join( group.set_index("Year")[["Department.Of.Labor.Cleaned.Low.Value.2020.Dollars"]].rename(columns={'Department.Of.Labor.Cleaned.Low.Value.2020.Dollars': name}))
    
min_wage.head()
```

```python
min_wage.describe()
```

Notemos que hay estados en los que todos los valores que tienen son ceros.

```python
min_wage.corr().head()
```

```python
issue_df = df[df['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars'] == 0]
issue_df.head()
```

```python
issue_df["State"].unique()
```

```python
for state, data in issue_df.groupby('State'):
    if data['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars'].sum() != 0:
        print("No es completamente vacía")
```

Todos los estados anteriores tienen valores indeseados, así que los quitaremos de nuestro dataset

```python
min_wage.replace(0, np.nan).dropna(axis=1).corr().head() # axis=1 para columnas, 0 para renglones
```

```python
min_wage_corr = min_wage.replace(0, np.nan).dropna(axis=1).corr()
```

```python
for problem in issue_df['State'].unique():
    if problem in min_wage_corr.columns:
        print("Falta algo por arreglar!")
```

Ahora podemos estar seguros de que la correlacion entre estados está libre de valores nulos y de los estados a los que les faltaban datos.

```python
plt.matshow(min_wage_corr, cmap = 'ocean')
plt.show()
```

Necesitamos cmabiar los valores de los ejes por las abreviaciones de los estados para que nuestro gráfico tenga sentido, si intentamos tomar las primeras dos letras de cada estado, obtendremos duplicados, como Mi(chigan) y Mi(nesota), obtendremos los datos de una página web y esas serán nuestras abreviaciones.

```python
import requests
web = requests.get("https://www.infoplease.com/state-abbreviations-and-state-postal-codes")
dfs = pd.read_html(web.text)
```

```python
for dataf in dfs:
    print(dataf.head())
```

```python
state_abbv = dfs[0]
state_abbv.head()
```

Si corremos el código nuevamente, la página podría negarnos el acceso y generar un error, así que es buena práctica guardar en un archivo la información obtenida.

```python
state_abbv.to_csv("./archive/state_abbv.csv", index = False)
```

```python
state_abbv = pd.read_csv("./archive/state_abbv.csv", index_col=0)
```

```python
abbv_dict = state_abbv[["Postal Code"]].to_dict()
abbv_dict = abbv_dict["Postal Code"]
abbv_dict
```

```python
labels = [abbv_dict[c] for c in min_wage_corr.columns]
```

Agregaremos de forma manual las abreviaciones faltantes.

```python
abbv_dict['Guam'] = "GU"
abbv_dict['Puerto Rico'] = "PR"
labels = [abbv_dict[c] for c in min_wage_corr.columns]
```

```python
fig, ax = plt.subplots(1,1, figsize=(10, 10))
cax = ax.matshow(min_wage_corr, cmap = 'ocean')

fig.colorbar(cax, shrink = 0.8)

ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))

ax.set_yticklabels(labels)
ax.set_xticklabels(labels)

plt.show()
```
